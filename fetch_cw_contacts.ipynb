```python
"""
Time:
pageSize = 100 (26min)
"""

import requests
import pandas as pd

url = "https://cw.emazzanti.net/v4_6_release/apis/3.0/company/contacts/"

headers = {
    'clientId': 'c1a21f3f-3b05-4b3f-95af-672d38fa465c',
    'Authorization': 'Basic ZW1henphbnRpK1dzZE5YeW12ZWdDWHlNVTg6bFM1T3BCb3F4a0RBVFZ0Yg=='
}

# Set initial parameters for the API request
params = {
    'conditions': 'email!=null and id!=null',
    'orderBy': 'id desc',  # Sort by id in descending order
    'pageSize': 1000,  # Adjust pageSize to fetch more records per request (e.g., 100)
    'page': 1  # Start with the first page
}

all_data = []  # To store all records

while True:
    print("Page: ", params['page'])
    response = requests.get(url, headers=headers, params=params)
    data = response.json()
    if not data:
        break
    all_data.extend(data)
    params['page'] += 1

df = pd.DataFrame(all_data)
df.to_excel("all_cw_contacts.xlsx", index=False)
```
